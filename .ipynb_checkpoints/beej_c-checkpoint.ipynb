#### STRUCT

```c
#include <stdio.h>
struct car {
    char *name;
    float price;
    int speed;
};

void set_price(struct car *c, float new_price) 
{
    // c.price = new_price; <<< this won't work, dot operator works only on structs
    (*c).price = new_price;
    // OR use arrow operator
    c->price = new_price;
}

int main(void) 
{
    struct car saturn;
    saturn.name = "Saturn SL/2";
    saturn.price = 15999.99;
    saturn.speed = 175;
    
    printf("Name: %s\n", saturn.name);
    
    set_price(&saturn, 5000.0);
    printf("New price: %f\n", saturn.price);
}
```

#### ADDING TO POINTERS

```c
#include <stdio.h>
int main(void)   
{
    int a[5] = {1,2,3,4};
    int *p = a;
    
    printf("%d\n", *p);
    printf("%d\n", *(p + 1));
    
    for (int i = 0; i < 5; i++) {
        printf("a[%d] = %d\n", i, *(p + i));
    }
}
```

#### CHANGING POINTERS

```c
#include <stdio.h>
int main(void)   
{
    int a[] = {1,2,3,4,9999};
    int *p = a;
    
    while(*p != 9999) {
        printf("%d\n", *p);
        p++;
    }
}
```

#### SUBTRACTING POINTERS

```c
#include <stdio.h>

int my_strlen(char *s)
{
    char *p = s;
    
    while(*p != '\0')
        p++;
    
    return p - s;
}

int main(void)   
{
    printf("%d\n", my_strlen("Hello World"));
}
```

#### VOID POINTER 

Void pointer cannot be dereferenced!!! You have to convert them to other types.

```c
#include <stdio.h>
#include <string.h>
int main(void)
{
    int a[] = {1,2,3};
    int b[3];
    
    memcpy(b, a, 3 * sizeof(int));
    
    printf("%d\n", b[1]);
}
```

```c
char a = 'X';  // A single char
    
void *p = &a;  // p points to the 'X'
char *q = p;   // q also points to the 'X'
    
printf("%c\n", *p);  // ERROR--cannot dereference void*!
printf("%c\n", *q);  // Prints "X"
```
